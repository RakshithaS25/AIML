# 4. Visualize above with Bar, Box, FacetGrid, Heatmap, Crosstab, Line plot Submission ( Python File with dataset)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Average No. of passengers (day-to-day basis)', data=data)
plt.title('Bar Plot of Earnings by Category')
plt.xlabel('Category')
plt.ylabel('Earnings')
plt.show()
```

```python
# 2. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Year', y='Average No. of passengers (day-to-day basis)', data=data)
plt.title('Box Plot of Earnings by Year')
plt.xlabel('Year')
plt.ylabel('Earnings')
plt.show()

```

```python
# 3. FacetGrid (Faceted Plot)
g = sns.FacetGrid(data, col='Average No. of daily trips', height=4)
g.map(sns.barplot, 'Year', 'Average No. of passengers (day-to-day basis)', ci=None)
g.set_axis_labels('Year', 'Average No. of passengers (day-to-day basis)')
g.set_titles('Average No. of daily trips: {col_name}')
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle('FacetGrid: Earnings by Year for Different Categories')
plt.show()
```

```python
# 4. Heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

```

```python
# 5. Crosstab and Heatmap
cross_tab = pd.crosstab(data['Average No. of daily trips'], data['Year'])
plt.figure(figsize=(8, 6))
sns.heatmap(cross_tab, annot=True, cmap='Blues')
plt.title('Crosstab Heatmap: Earnings by Category and Year')
plt.show()

```

```python
# 6. Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Average No. of passengers (day-to-day basis)', hue='Average No. of daily trips', data=data)
plt.title('Line Plot of Earnings Over Years by Category')
plt.xlabel('Year')
plt.ylabel('Average No. of passengers (day-to-day basis)')
plt.legend(title='Average No. of daily trips')
plt.show()
```
