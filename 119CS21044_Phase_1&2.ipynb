# 6 Test type (T-test, Z-test, F-test, ANNOVA, Chi-Square, PCA)

```python
#T-Test:

import scipy.stats as stats

# Sample data for two groups
group1 = [10, 12, 15, 17, 20]
group2 = [14, 16, 18, 20, 22]

# Perform a two-sample independent t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")
```

```python
# Z-test:

import scipy.stats as stats

# Sample data (replace with your data)
sample_data = [18.23, 18.22, 18.40]
population_mean = 18.0
population_stddev = 0.5

# Calculate the Z-statistic
sample_mean = sum(sample_data) / len(sample_data)
z_statistic = (sample_mean - population_mean) / (population_stddev / (len(sample_data)**0.5))

# Set the significance level (alpha)
alpha = 0.05

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

# Check if the null hypothesis is rejected
if p_value < alpha:
    print(f"Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print(f"Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")

print(f"Z-Statistic: {z_statistic}")
print(f"P-Value: {p_value}")
```

```python
#F-Test:

import scipy.stats as stats

# Sample data from three groups
group1 = [12, 15, 18, 21, 24]
group2 = [14, 16, 19, 22, 25]
group3 = [10, 13, 16, 19, 22]

# Perform a one-way ANOVA
f_stat, p_value = stats.f_oneway(group1, group2, group3)

print(f"F-Statistic: {f_stat}")
print(f"P-Value: {p_value}")
```

```python
# Annova test
from scipy.stats import f_oneway

# Example: Test if there's a significant difference in average number of daily trips between different cities
agra = [245.0, 258.0, 217.0]
aizawl = [2.5, 2.5]
visakhapatnam = [7667.0, 7953.0]
warangal = [5377.0, 6102.0, 6022.0]

f_stat, p_value = f_oneway(agra, aizawl, visakhapatnam, warangal)
print(f"F-Stat: {f_stat}, P-Value: {p_value}")
```

```python
#Chi-Square Test:

import scipy.stats as stats

# Create a contingency table
observed = [[15, 10], [20, 25]]

# Perform a chi-squared test of independence
chi2, p, dof, expected = stats.chi2_contingency(observed)

print(f"Chi-Squared Statistic: {chi2}")
print(f"P-Value: {p}")
```

Interpretation:

Alternative Hypothesis (Ha): At least one city's daily trips mean is different from the others.
    
Interpretation of results:

→ If the p-value is less than the significance level (alpha), typically 0.05, you reject the null hypothesis.
This means there is enough evidence to conclude that there is a significant difference in daily trips among cities.
→ If the p-value is greater than alpha, you fail to reject the null hypothesis. This suggests that daily trips do
not significantly differ among cities.
